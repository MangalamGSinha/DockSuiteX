## 1. Molecule Fetching

DockSuiteX can download structures by ID from RCSB PDB, PubChem, and ChEMBL. It supports both single-molecule and batch downloads.

```python
# Batch download proteins (PDB)
protein_files = fetch_pdb(pdbid=["1M17", "1HVR"], save_to="data/proteins")

# Batch download ligands from different sources (PubChem CID and ChEMBL ID)
ligand_files = fetch_sdf(molecule_id=["2244", "CHEMBL25"], save_to="data/ligands")

print(f"Proteins: {protein_files}")
print(f"Ligands: {ligand_files}")
```

## 2. Format Conversion (Open Babel)

The `convert` utility is a powerful wrapper for Open Babel, supporting single files, lists, or entire directories.

```python
# 1. Convert a single file (e.g., SDF to PDB)
pdb_file = convert(input="data/ligands/2244.sdf", output_format="pdb")

# 2. Convert an entire directory of SDFs to PDBQT for docking
pdbqt_files = convert(input="data/ligands", output_format="pdbqt", output_dir="data/pdbqt_ligands")

# 3. Split a multi-model PDB into individual residue/model files
models = convert(input="multi_model.pdb", output_format="pdb", split=True)

# 4. Check what formats are supported locally
formats = get_supported_formats()
print(f"Supported formats: {len(formats)}")
print(f"Example: {formats['sdf']}")
```
